```python
Original_data = pd.read_csv("C:\\Users\\Desktop\\PowerBI\\Original_data.csv")
Original_data.rename(columns={'GENDER': 'gender'}, inplace=True)
```

```python
Finance_data = pd.read_csv("C:\\Users\\Desktop\\PowerBI\\Finance_data.csv")
```

<span style="color: blue; font-size: 20px; font-weight: bold;">Average Rank for Investment Options by Gender</span>

```python
investment_columns = ['Mutual_Funds', 'Equity_Market', 'Debentures', 'Government_Bonds', 'Fixed_Deposits', 'PPF', 'Gold']
data_avg = Finance_data.groupby('gender')[investment_columns].mean()

data_avg.plot(kind='bar', figsize=(10, 6), colormap='coolwarm')
plt.title('Average Investment Ranks by Gender')
plt.ylabel('Average Rank')
plt.xlabel('Gender')
plt.xticks(rotation=0)
plt.show()
```

<span style="color: blue; font-size: 20px; font-weight: bold;">Preferred Investment Duration</span>


```python
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)

sns.countplot(x='Duration', data=Finance_data, palette='viridis')
plt.title('Preferred Investment Duration')
plt.xlabel('Investment Duration')
plt.ylabel('Count')
plt.show()
```

<span style="color: blue; font-size: 20px; font-weight: bold;">Reasons for Investing in Specific Avenues</span>


```python
sns.countplot(data=Finance_data, x='Reason_Equity', palette='YlGnBu')
plt.title('Reasons for Investing in Equity')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()

sns.countplot(data=Finance_data, x='Reason_Mutual', palette='YlOrRd')
plt.title('Reasons for Investing in Mutual Funds')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

<span style="color: blue; font-size: 20px; font-weight: bold;">Most Popular Investment Option</span>


```python
investment_types = ['Mutual_Funds','Equity_Market','Debentures','Government_Bonds','Fixed_Deposits','PPF','Gold']
investment_popularity = Finance_data[investment_types].sum()

investment_popularity.plot(kind='bar', color='lightgreen', figsize=(12, 6))
plt.title('Most Popular Investment Avenues')
plt.ylabel('Count of People')
plt.xticks(rotation=0)
plt.show()
```

<span style="color: blue; font-size: 20px; font-weight: bold;">Most Common Sources of Information for Investment Decisions
</span>


```python
source_info = Finance_data['Source'].value_counts()

source_info.plot(kind='bar', color='lightblue', figsize=(10, 6))
plt.title('Sources of Information for Investment')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```
